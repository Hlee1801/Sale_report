```python
import pandas as pd
import os 
import matplotlib.pyplot as plt  
 
```

task 1: import packages and load data 

```python
path = 'E:\Mini project\Sales-Reporting\data\\'
dataframe = pd.read_csv(path +'sales2019_1.csv')
dataframe.head() 
```

Task 2: clean and preprocess data from
task 2.1: merge 12-month data from

```python
frame = []
all_length = []
for file in os.listdir(path):
    if file.endswith('.csv'):
        filepath = path + file
        df1 = pd.read_csv(filepath)
        frame.append(df1) 
        results = pd.concat(frame)
        length_1month = len(df1.index)
        all_length.append(length_1month)
df = results

df.to_csv('Tong san pham ban duoc 2019')
```

```python
df.head()
```

Task 2.2 add 'month' column 

```python
df['month'] = df['Order Date'].str[0:2] 
df.head()
```

Task 2.3 get rid of 'NaN' and 'Or' value 

```python
df = df[df['month'] != 'Or']
df = df.dropna(how = 'all')
df.head()
```

Task 3 : Reporting
Task 3.1: what was the best month for sales? how much was earned that month? 

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], downcast = 'integer')
df['Price Each'] = pd.to_numeric(df['Price Each'], downcast = 'float')
```

```python
df['sale'] = df['Quantity Ordered']* df['Price Each']
df.head()
```

```python
sale_value = df.groupby('month').sum()['sale']
sale_value.max
```

```python
months = range(1,13)
plt.bar(x = months, height = sale_value)
plt.xticks(months)
plt.xlabel('Months')
plt.ylabel('Sales in USD')
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Hours'] = df['Order Date'].dt.hour
sales_value_hours = df.groupby('Hours').count()['sale']
hours = [hour for hour, sale in sales_value_hours.items()]
plt.plot(hours, sales_value_hours)
plt.grid()
plt.xticks(hours, rotation=90, size=8)
plt.xlabel('Hours')
plt.ylabel('Sales in USD')
plt.show()
```

```python
df_dup = df[df['Order ID'].duplicated(keep=False)]
groupProduct = lambda product: ', '.join(product)
df_dup['All Products'] = df_dup.groupby('Order ID')['Product'].transform(groupProduct)
df_dup = df_dup[['Order ID', 'All Products']].drop_duplicates()
df_dup['All Products'].value_counts().head(10)
```

```python
all_products = df.groupby('Product').sum()['Quantity Ordered']
prices = df.groupby('Product').mean()['Price Each']
products_ls = [product for product, quant in all_products.items()]


x = products_ls
y1 = all_products
y2 = prices

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(x, y1, color='g')
ax2.plot(x, y2, 'b-')

ax1.set_xticklabels(products_ls, rotation=90, size=8)
ax1.set_xlabel('Products')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price Each', color='b')

plt.show()
```
