```python
import pandas as pd 
import os 
```

```python
path = 'E:\Mini project\Sales-Reporting\data\\'
dataframe = pd.read_csv(path+'sales2019_1.csv')
dataframe.head(30)
```

```python
frames = []
all_length = []
for file in os.listdir(path):
    if file.endswith('.csv'):
        filepath = path + file 
        df1 = pd.read_csv(filepath)
        frames.append(df1)
        result = pd.concat(frames)
        length_1month = len(df1.index)
        all_length.append(length_1month)
df = result
df.to_csv('TotalSales2019.csv',index = False )
```
